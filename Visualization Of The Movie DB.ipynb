# **Data of Movies Which Has Rating Less Than 50**

```python
df = dataset[dataset['Rating'] < 50]
```

```python
df
```

```python
least_20 = df[['Name','Rating']]
least_20
```

# Data Of Least 20 Movies Based On Ratings

```python
least1_20 = least_20.sort_values(by='Rating', ascending=True)
least1_20 = least1_20[30:51]
least1_20
```

# Least 20 Movies Based On Ratings

```python
import matplotlib.pyplot as plt
fig = plt.figure()
fig = plt.figure(figsize=(10,7))
sns.barplot(x = 'Name', y = 'Rating', data = least1_20).set(title = 'Least 20 Movies Based On The Ratings')
plt.xticks(rotation = 90)
plt.show()
```
