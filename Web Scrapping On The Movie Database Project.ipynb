# WebScrapping Project Of The Movie ORG DB #

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.themoviedb.org'
```

```python
header = {'User-Agent':'Mozilla/5.0(Windows NT 6.3;Win64;x64)AppleWebKit/537.36(KHTML,like Gecko)Chrome/99.04844.51 Safari/537.36'}
url = 'https://www.themoviedb.org'

response = requests.get(url+'/movie', headers = header)
```

```python
response
```

```python
movie_data = response.text
```

```python
Soup = BeautifulSoup(movie_data , 'lxml')
```

# Scrapping First Page Data ###

```python
fpage_movies_data = []
fpage_link = []
fpg_url = []
fpage_all_mov = Soup.find_all('div' , class_='card style_1')
for items in fpage_all_mov:
    mov_name = items.find('div' , class_='content').h2.text
    mov_rat = items.find('div' , class_='user_score_chart')['data-percent']
    movie_link = items.find('div',class_="content").a['href']
    mov_link = 'https://www.themoviedb.org' + movie_link
    fpage_link.append(mov_link)
    
    fpage_dct1 = {
        'Name':mov_name,
        'Rating':mov_rat
    }
    fpage_movies_data.append(fpage_dct1)
    
    url1 = {
        'Url':mov_link
    }
    fpg_url.append(url1)
```

```python
fpg_data = pd.DataFrame(fpage_movies_data)
fpg_url = pd.DataFrame(fpg_url)
```

# Fetching 20 Movie Links Of First Page

```python
for i in fpage_link:
    print(i)
```

```python
fpg_facts = []
def pgfct():
    for i in fpage_link:
        header = {'User-Agent':'Mozila/5.0(Windows NT 6.3;Win64;x64)AppleWebkit/537.36(KHTML,like Gecko)Chrome/99.0.4844.51 Safari/537.36'}
        dt = requests.get(i , headers = header)
        fcn = dt.text
        fsp = BeautifulSoup(fcn , 'lxml')
        #print(fsp)
        ffcts = fsp.find_all('section' , class_='header poster')
        for d in ffcts:
            fr = d.find('span' , class_='release').text.strip()[0:10]
            fgn = d.find('span' , class_='genres').text.replace('\n' , '')
            frt = d.find('span' , class_='runtime').text.replace('\n' , '')
            dr = d.find_all('li' , class_='profile')
            fdr = [dn.get_text() for dn in dr]
            for dtd in fdr:
                if 'Director' in dtd:
                    ddfp = dtd.partition('Director')[0]
                    #print(ddfp)
            fpg_dct = {
                'Genre':fgn,
                'Release date':fr,
                'Runtime':frt,
                'Director':ddfp.replace('\n' , '')
            }
            #print(fpg_dct)
            fpg_facts.append(fpg_dct)
pgfct()
    
```

```python
ffct_data = pd.DataFrame(fpg_facts)
#print(ffct_data)
```

# DataFrame of 1st Page Data ####

```python
fpage1 = fpg_data.join(ffct_data)
fpagef = fpage1.join(fpg_url)
fpagef
```

# Fetching Movies Links Of 150 Pages ###

```python
url_lst=[]
url="https://www.themoviedb.org/movie?page="
for i in range(1,151):
    nw_link=url+str(i)
    url_lst.append(nw_link)
url_lst
```

```python
for lnk in url_lst:
    print(lnk)
```

# Link of all Movies of 150 Pages ###

```python
allmvlk = []
for lnk in url_lst:
    h1 = {'User-Agent':'Mozila/5.0(Windows NT 6.3;Win64;x64)AppleWebkit/537.36(KHTML,like Gecko)Chrome/99.0.4844.51 Safari/537.36'}
    l_dt = requests.get(lnk , headers = h1)
    lct = l_dt.text
    lsp = BeautifulSoup(lct , 'lxml')
    ldt = lsp.find_all('div' , class_='card style_1')
    for dl in ldt:
        allmov_lk = dl.find('a',class_='image')['href']
        allmovie_links = 'https://www.themoviedb.org'+allmov_lk
        allmvlk.append(allmovie_links)
```

```python
allmvlnk_dct = pd.DataFrame(allmvlk)
allmvlnk_dct
```

# Movie Data of 150 Pages ###

```python
allpage_mov_data = []
allUrl = []
def allfct():
    for lnk in url_lst:
        h2 = {'User-Agent':'Mozila/5.0(Windows NT 6.3;Win64;x64)AppleWebkit/537.36(KHTML,like Gecko)Chrome/99.0.4844.51 Safari/537.36'}
        all_dt = requests.get(lnk , headers = h2)
        all_ct = all_dt.text
        all_sp = BeautifulSoup(all_ct , 'lxml')
        alldt = all_sp.find_all('div' , class_='card style_1')
        for item in alldt:
            allmov_name = item.find('div' , class_='content').h2.text
            allmov_rat = item.find('div' , class_='user_score_chart')['data-percent']
            allmov_url = item.find('a',class_='image')['href']
            all_links = 'https://www.themoviedb.org' + allmov_url
        
            all_data = {
             'Name':allmov_name,
             'Rating':allmov_rat
            }
            allpage_mov_data.append(all_data)
            #print(allpage_mov_data)
        
            allurl = {
             'Url':all_links
             }
            allUrl.append(allurl)
allfct()
```

```python
alldtd = pd.DataFrame(allpage_mov_data)
allurl_dt = pd.DataFrame(allUrl)
alldtd
```

# Scrapping All Data Of The Website

```python
allpg_fcts = []
alldr = []
allrt = []
def allpgft():
    for all_lnk in allmvlk:
        h3 = {'User-Agent':'Mozila/5.0(Windows NT 6.3;Win64;x64)AppleWebkit/537.36(KHTML,like Gecko)Chrome/99.0.4844.51 Safari/537.36'}
        ftdt = requests.get(all_lnk , headers = h3)
        ftct = ftdt.text
        ftsp = BeautifulSoup(ftct , 'lxml')
        allpg_dt = ftsp.find_all('div' , class_='header large border first')
        for items in allpg_dt:
            allpg_gnr = items.find('span' , class_='genres')
            if allpg_gnr is not None:
                allpg_gnr = items.find('span' , class_='genres').text.replace('\n' , '')
            else:
                allpg_gnr = 'None'
            allpg_rls = items.find('span' , class_='release')
            if allpg_rls is not None:
                allpg_rls = items.find('span' , class_='release').text.strip()[0:10]
            else:
                allpg_rls = 'None'
            allpg_rt = items.find('span' , class_='runtime')
            if allpg_rt is not None:
                allpg_rt = items.find('span' , class_='runtime').text.replace('\n' , '')
            else:
                allpg_rt = 'None'
            #print(allpg_rt)
            allpg_dr = items.find_all('li' , class_='profile')
            #print(allpg_dr)
            allpgdn = [aldn.get_text() for aldn in allpg_dr]
            for alldnd in allpgdn:
                if 'Director' in alldnd:
                    if alldnd is not None:
                        alldnd1 = alldnd.partition('Director')[0]
                        if 'Characters,' in alldnd1:
                            alldnd2 = alldnd1.partition('Characters,')[0]
                        else:
                            alldnd2 = alldnd1
                    else:
                        alldnd1 = 'None'
                    #alldr.append(aldr)
            allpg_dct = {
                'Genre':allpg_gnr,
                'Release date':allpg_rls,
                'Runtime':allpg_rt,
                'Director':alldnd2.replace('\n' , '')
            }
            #print(allpg_dct)
            #allpg_fcts.append(allpg_dct)
            
#allpgft()
```
